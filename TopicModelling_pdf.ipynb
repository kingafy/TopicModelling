```python
##DATA INGESTION and CONSOLIDATION
import re
from __future__ import absolute_import, division, print_function
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage
##from cStringIO import StringIO
import io
```

```python
def convert_pdf_to_txt(path):
    rsrcmgr = PDFResourceManager()
    ##retstr = StringIO()
    retstr = io.StringIO()
    codec = 'utf-8'
    laparams = LAParams()
    device = TextConverter(rsrcmgr, retstr, codec=codec, laparams=laparams)
    fp = open(path, 'rb')
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    password = ""
    maxpages = 0
    caching = True
    pagenos=set()

    for page in PDFPage.get_pages(fp, pagenos, maxpages=maxpages, password=password,caching=caching, check_extractable=True):
        interpreter.process_page(page)

    text = retstr.getvalue()

    fp.close()
    device.close()
    retstr.close()
    return text
```

```python
##Data cleaning..Remove double quotes
shear=[i.replace('\xe2\x80\x9c','') for i in clean_cont ]
shear=[i.replace('\xe2\x80\x9d','') for i in shear ]
shear=[i.replace('\xe2\x80\x99s','') for i in shear ]
```

```python
shears = [x for x in shear if x != ' ']
shearss = [x for x in shears if x != '']
```

```python
dubby=[re.sub("[^a-zA-Z]+", " ", s) for s in shearss]
```

```python
####POST CLEANING ACTIVITIES
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import pandas as pd
import numpy as np
%pylab
%matplotlib inline
```

```python
from sklearn.feature_extraction.stop_words import ENGLISH_STOP_WORDS
```

```python
vect=CountVectorizer(ngram_range=(1,1),stop_words='english')
```

```python
dtm=vect.fit_transform(dubby)
```

```python
##Create dataframe of the Dtm
pd.DataFrame(dtm.toarray(),columns=vect.get_feature_names())
```

```python
##Running LDA on the word matrix
lda=LatentDirichletAllocation(n_components=5)
```

```python
##Fit the data to lda
lda.fit_transform(dtm)
```
